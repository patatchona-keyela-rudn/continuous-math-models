```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# PARAMETERS
# N0 - initial population
# L - medium(container) capacity
# alpha - multiplication speed

# GOMPERZ MODEL
def N_gomperz(t,N0,L,alpha):
    return L * np.power(N0/L, np.exp(-1.*alpha*t / np.log(L)))

# FERHULST MODEL
def N_ferhulst(t,N0,L,alpha):
    return L*N0 / (N0 + (L-N0) * np.exp(-1.0*alpha*t))

# DERIVATIVES OF THE FUNCTIONS
def dN_gomperz(t,N0):
    return -1.0 * np.exp(-0.1*t/np.log(10)) * np.power(N0/10,np.exp(-0.1*t/np.log(10))) * np.log(0.2)/np.log(10)

def dN_ferhulst(t,N0):
    return (10-N0) * N0 * np.exp(-0.1*t) / ((N0 + (10-N0) * np.exp(-0.1*t))**2)
```

```python
t = np.linspace(0,100,1000)
```

```python
# Different values of N0
N0s = [2,4,6]
for N0 in N0s:
    dN = dN_gomperz(t,N0)
    max_index = np.argmax(dN)
    
    plt.figure(figsize=(8,8/1.618))
    plt.plot(t,dN,label="   ")
    plt.scatter(t[max_index],dN[max_index],c='r',marker='x',label="  ")
    plt.plot([t[max_index],t[max_index]],[0,dN[max_index]],'r--')
    
    plt.xlim(-1,100)
    plt.ylim(0,max(dN)*1.1)
    if t[max_index] != 0:
        plt.xticks([0,20,40,60,80,100,t[max_index]])
    
    plt.legend()
    plt.xlabel("time,$t$")
    plt.ylabel("$N'(t)$")
    plt.title(f" , $N_0$ = {N0}")
    
    plt.tight_layout()
    plt.savefig(f"dN_gomperz_N0_{N0}.png",dpi=300)
    plt.show()
    
    N = N_gomperz(t,N0,L=10,alpha=0.1)
    
    plt.figure(figsize=(8,8/1.618))
    plt.plot(t,N,label="   ")
    plt.plot([0,100],[10,10],'k--')
    plt.scatter(t[max_index],N[max_index],c='r',marker='x',label="  ")
    plt.plot([t[max_index],t[max_index]],[0,N[max_index]],'r--')
    
    plt.xlim(-1,100)
    plt.ylim(0,max(N)*1.1)
    if t[max_index] != 0:
        plt.xticks([0,20,40,60,80,100,t[max_index]])
    
    plt.legend()
    plt.xlabel("time,$t$")
    plt.ylabel("$N(t)$")
    plt.title(f" , $N_0$ = {N0}")
    
    plt.tight_layout()
    
    plt.savefig(f"N_gomperz_N0_{N0}.png",dpi=300)
    plt.show()
    
```

```python
# Different values of N0
N0s = [2,4,6]
for N0 in N0s:
    dN = dN_ferhulst(t,N0)
    max_index = np.argmax(dN)
    
    plt.figure(figsize=(8,8/1.618))
    plt.plot(t,dN,label="   ")
    plt.scatter(t[max_index],dN[max_index],c='r',marker='x',label="  ")
    plt.plot([t[max_index],t[max_index]],[0,dN[max_index]],'r--')
    
    plt.xlim(-1,100)
    #plt.ylim(0,max(dN)*1.1)
    if t[max_index] != 0:
        plt.xticks([0,20,40,60,80,100,t[max_index]])
    
    plt.legend()
    plt.xlabel("time,$t$")
    plt.ylabel("$N'(t)$")
    plt.title(f" , $N_0$ = {N0}")
    
    plt.tight_layout()
    plt.savefig(f"dN_ferhulst_N0_{N0}.png",dpi=300)
    plt.show()
    
    N = N_ferhulst(t,N0,L=10,alpha=0.1)
    
    plt.figure(figsize=(8,8/1.618))
    plt.plot(t,N,label="   ")
    plt.plot([0,100],[10,10],'k--')
    plt.scatter(t[max_index],N[max_index],c='r',marker='x',label="  ")
    plt.plot([t[max_index],t[max_index]],[0,N[max_index]],'r--')
    
    plt.xlim(-1,100)
    plt.ylim(0,max(N)*1.1)
    if t[max_index] != 0:
        plt.xticks([0,20,40,60,80,100,t[max_index]])
    
    plt.legend()
    plt.xlabel("time,$t$")
    plt.ylabel("$N(t)$")
    plt.title(f" , $N_0$ = {N0}")
    
    plt.tight_layout()
    
    plt.savefig(f"N_ferhulst_N0_{N0}.png",dpi=300)
    plt.show()
    
```

```python
index_max_gomperz,index_max_ferhulst = [], []
for N0 in N0s:
    dN = dN_gomperz(t,N0)
    max_index = np.argmax(dN)
    index_max_gomperz.append(max_index)
    
    dN = dN_ferhulst(t,N0)
    max_index = np.argmax(dN)
    index_max_ferhulst.append(max_index)
```

```python
for i, N0 in enumerate(N0s):
    plt.figure(figsize=(8,8/1.618))
    
    N1 = N_gomperz(t,N0,L=10,alpha=0.1)
    N2 = N_ferhulst(t,N0,L=10,alpha=0.1)
    id1, id2 = index_max_gomperz[i], index_max_ferhulst[i]
    
    plt.plot(t,N1,'m-',label=" ")
    plt.plot(t,N2,'c-',label=" ")
    plt.plot([0,100],[10,10],'k--')
    
    plt.scatter(t[id1],N1[id1],c='r',marker='x')
    plt.scatter(t[id2],N2[id2],c='b',marker='x')
    
    plt.plot([t[id1],t[id1]],[0,N1[id1]],'r--')
    plt.plot([t[id2],t[id2]],[0,N2[id2]],'b--')
    
    plt.xlim(-1,100)
    plt.ylim(N0-0.25,10.25)
    
    plt.legend()
    plt.xlabel("time,$t$")
    plt.ylabel("$N(t)$")
    plt.title(f"   $N_0$ = {N0}")
    
    plt.tight_layout()
    
    plt.savefig(f"compare_N0_{N0}.png",dpi=300)
    plt.show()
```
